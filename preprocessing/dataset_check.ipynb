```python
import pandas as pd

csv_path = "/home/ubuntu/datasets/CICDDoS2019/CSV-03-11/03-11/UDP.csv"
df = pd.read_csv(csv_path)

# Remove leading/trailing spaces from column names
df.columns = [c.strip() for c in df.columns]

# Specific 5-tuple to search for
src_ip = "172.16.0.5"
dst_ip = "192.168.50.4"
src_port = 32768
dst_port = 31143
protocol = 17

matches = df[
    (df["Source IP"].astype(str) == src_ip) &
    (df["Destination IP"].astype(str) == dst_ip) &
    (df["Source Port"].astype(int) == src_port) &
    (df["Destination Port"].astype(int) == dst_port) &
    (df["Protocol"].astype(int) == protocol)
]

print("Matches found:", len(matches))

matches[
    [
        "Flow ID",
        "Source IP",
        "Source Port",
        "Destination IP",
        "Destination Port",
        "Protocol",
        "Timestamp",
        "Label",
    ]
].head(20)
```
